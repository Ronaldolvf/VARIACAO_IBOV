```python
# Importando as bibliotecas que serão utilizadas. 
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import datetime as dt
import time
```

```python
tempo_inicial = time.time()

# Importando os dados do aquivo "IBOVDiA3.csv".
dados = pd.read_csv('dados/IBOVDia_12-09-22.csv')   

# Selecinando os tickers das ações.
tickers = dados["CD"]
tickers
```

```python
# Informando a data inicial e final.
data_inicial = dt.datetime(2022,9,9)
data_final   = dt.datetime(2022,9,12)

# Criando uma variável para os dados extraídos.
lista = tickers

# Importando as cotações das ações (tickers) via "yahoo finance,".
df = wb.DataReader (lista, "yahoo", data_inicial, data_final) 
df.round(2)
```

```python
# Usando a função "pct_change() " para calcular a variação diária.
df['Close'] = df['Close'].pct_change()         

# Selecionando a coluna "Close" e multiplicando por 100 as cotações da linha 01.
# Colocando 2 colchetes, cria-se um data-frame. Essa função ocorre graças ao pandas. 
test = df[['Close']].round(4)*100    
test           
```

```python
# Selecionando os dados da linha 01.
ibov = test.iloc[1].round(3)
ibov

# Transformando os dados em um "DataFrame"
ibov = pd.DataFrame (ibov)

# Ordenando os dados conforme a data selecionada.
ibov.sort_values(['2022-09-12'], ascending = False, inplace = True)
ibov
```

```python
ibov.reset_index(inplace = True)
ibov.drop(["Attributes"], axis=1, inplace = True)
ibov.columns = ["Symbols", "variation"]
ibov
```

```python
def checar_diff_acao(valor):
    if valor > 0:  
        #color =  f'\033[92m up \033[0m '
        return "up"    
    elif valor < 0:
        return "down"
    else:
        return "egual"
    
```

```python
# Atribuir rotulos
ibov["State.variation"] = ibov["variation"].apply(checar_diff_acao) 
ibov
```

```python
# Exportando via arquivo xlsx.
ibov.to_excel('ibov_cotação.xlsx')
```

```python
tempo_final = time.time()

tempo = int(tempo_final - tempo_inicial)/60
tempo
```
